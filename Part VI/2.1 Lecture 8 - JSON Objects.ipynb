# JSON Objects


## Rules 
 * Object structures begins with curly brackets `{}`
 * Data is contained within the curly brackets, as key-value pairs
 * Each key-value pair has to be separated by a colon `:`
 * Each Key must be a string encased in double quotes `"`
 * Values can take many forms (see below)
 * If there are multiple key value pairs, you must put a comma after each pairing until the last pairing. 
 
 ## Value Data Types
* a string (like the examples above) 
* a number
* true or false
* null
* another object
* an array  
 
 ## Examples
 
 ### Basic Example

```JSON
{
    "Key":"value"
}
```

### With Multiple Key-Value Pairs
```JSON
{
    "key":"value",
    "key2":"value2"
}
```

### Where Values are Strings
```JSON
{
"Item":"Hoodie"
}
```

### Where Values are Numbers
```JSON
{
"Year":1953
}
```

### Where Values are True or False
```JSON
{
"isActive":true,
"isReady":false
}
```

### Where Values are Null
```JSON
{
"price":null
}
```

### Where Value is another Object
```JSON
{
"key":{
        "key":"value"
      }
}
```

### Where value is an Array
```JSON
{
"key":[]
}
```
