# JSON Arrays


## Rules
* Arrays begin with straight brackets `[]`
* Values sit within the brackets
* If there are multiple values, you must put a comma after each value until the last value

## Value Data Types
Data Types are exactly the same for Arrays as they are for Objects. As a review, values can be any of the following: 
* a string
* a number
* true or false
* null
* an object
* another array  


## Examples

### Where Values are Strings

```JSON
[
    "Bananas",
    "Eggs",
    "Milk"
]
```


### Where Values are Numbers

```JSON
[
    1000,
    1001,
    1002,
    1003
]
```

### Where Values are True or False 

```JSON 
[
    true,
    false,
    false,
    true
]
```

### Where Values are Null

```JSON 
[
    null,
    12,
    null,
    null
]
```

### Where Values are Objects

```JSON
[
    {"key":"values"},
    {"key":"values"},
    {"key":"values"}

]
```

### Where Values are Arrays

```JSON
[
    ["Milk","Eggs","Cheese"],
    ["Ham","Pineapple"]
]

```
